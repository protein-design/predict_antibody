```python
import numpy as np
import pandas as pd

import matplotlib as mpl
print(mpl.__version__)
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec

mpl.rcParams['font.family'] = 'sans-serif'
mpl.rcParams['font.sans-serif'] = ['Arial', 'DejaVu Sans', 'sans-serif']

%load_ext autoreload
%autoreload 2
    
from layout import Layout
```

```python
def plot1(ax):
    x = np.linspace(0, 10, 100)
    y = np.sin(x)
    ax.plot(x, y, color='red')
    ax.set_xlabel('x')
    ax.set_ylabel('y')

def hist1(ax):
    x = np.linspace(0, 10, 100)
    y = np.sin(x)
    ax.hist(y)
    ax.set_xlabel('counts')
    ax.set_ylabel('values')
```

```python
# grid: 1st row one figure, 2nd row two figures
args = {'width_level': 3, 'height': 10, 'wspace':1.4, 'hspace':.8}
label_xytext = [(-40,16), (-40,16), (-40,16), (-40,16), (-40,16)]
fig, axes = Layout(args).row2_row3(label_xytext)
plot1(axes[0])
hist1(axes[1])
plot1(axes[2])
```

```python
fig, ax = Layout().one()
plot1(ax)
```

```python
args = {'width_level': 3, 'height': 10, 'space':.1}
label_xytext = [(-35,12), (-35,12), (-35,12), (-35,12),]
fig, axes = Layout(args).grid_22(label_xytext)
plot1(axes[0])
```

```python
# two columns in one row
args = {'width_level': 3, 'height': 4}
label_x = (-30, -30)
label_y = 16
fig, axes = Layout(args).row(label_x, label_y)
```

```python
# three columns in one row
args = {'width_level': 3, 'height': 4}
label_x = (-30, -30, -30)
label_y = 16
fig, axes = Layout(args).row(label_x, label_y)
```

```python
# two rows in one column
args = {'width_level': 1, 'height': 8}
label_x = -30
label_y = (16,16)
fig, axes = Layout(args).column(label_x, label_y)
plot1(axes[0])
```

```python
# grid: 1st row one figure, 2nd row two figures
args = {'width_level': 3, 'height': 8}
label_xytext = [(-40,16), (-40,16), (-40,16)]
fig, axes = Layout(args).row1_row2(label_xytext)
plot1(axes[0])
hist1(axes[1])
plot1(axes[2])
fig.savefig('demo.tif', dpi=600, bbox_inches='tight', pad_inches=.05)
```
